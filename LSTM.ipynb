```python
def preprocess_data(data, scaling_method, min_max_range=None, zero_mean_range=None):
    data = data.astype('float32')
    if scaling_method == "min-max":
        if min_max_range is None:
            scaler = MinMaxScaler(feature_range=(0, 1))
        else:
            scaler = MinMaxScaler(feature_range=min_max_range)
    elif scaling_method == "zero-mean":
        if zero_mean_range is None:
            scaler = StandardScaler()
        else:
            zero_mean_range = bool(zero_mean_range)
            scaler = StandardScaler(with_mean=zero_mean_range)
    elif scaling_method == "ch∆∞a chu·∫©n h√≥a":
        scaler = None  
    else:
        raise ValueError("Ph∆∞∆°ng ph√°p chu·∫©n h√≥a kh√¥ng h·ª£p l·ªá. Ch·ªçn 'zero-min', 'min-max' ho·∫∑c 'ch∆∞a chu·∫©n h√≥a'.")
    if scaler is not None:
        scaled_data = scaler.fit_transform(data)
    else:
        scaled_data = data
    return data, scaled_data, scaler
```

```python
def train_test_split(data, train_size, validation_size, test_size):
    total_size = len(data)
    train_size = int(total_size * train_size)
    validation_size = int(total_size * validation_size)
    test_size = total_size - train_size - validation_size
    train, validation, test = (
        data[:train_size, :],
        data[train_size:train_size + validation_size, :],
        data[train_size + validation_size:, :]
    )
    return train, validation, test,train_size,validation_size
train_size = 0.6
validation_size = 0.2
test_size = 0.2
train, validation, test , train_size, validation_size= train_test_split(data, train_size, validation_size, test_size)
train
print(test.shape)
```

# <center> <span style="font-family:Playfair Display"> <span style="color:white;display:fill;border-radius:8px;background-color:#a02933;font-size:100%;font-weight: bold;color: #dbba78;font-style: italic;">„Ö§üî¢ T·∫°o ma tr·∫≠n d·ªØ li·ªáu ƒë·∫ßu v√†o v√† ƒë·∫ßu ra cho c√°c m√¥ h√¨nh time series LSTM g·ªìm  </span><a id="1"></a>  
* x l√† c√°c gi√° tr·ªã ƒë·∫ßu v√†o t·ª©c T-1 , T-2 ,T-lookback
* y l√† c√°c gi√° tr·ªã ƒë·∫ßu ra t·ª©c T, T+1 , T+ look_forward </br>
</br>


```python

def create_dataset(dataset, look_back=1, look_forward=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back - look_forward + 1):
        # Tr√≠ch xu·∫•t d√£y ƒë·∫ßu v√†o v√† gi√° tr·ªã m·ª•c ti√™u
        seq_x = dataset[i:(i + look_back), 0] #T-1,T-2,T-3,..T-lookback
        target_y = dataset[i + look_back:i + look_back + look_forward, 0] #T,T+1,..,T+look_forward
        # Ki·ªÉm tra gi√° tr·ªã NaN trong d√£y ƒë·∫ßu v√†o v√† gi√° tr·ªã m·ª•c ti√™u
        if not np.isnan(seq_x).any() and not np.isnan(target_y).any():
            dataX.append(seq_x)
            dataY.append(target_y)
    return np.array(dataX), np.array(dataY)
```

```python

def create_rnn_model(input_shape, output_units, units=62):
    model = Sequential()
    model.add(SimpleRNN(units, input_shape=input_shape, return_sequences=True)) 
    model.add(Dropout(0.2))
    model.add(SimpleRNN(units, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(output_units))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
model_rnn = create_rnn_model #T·∫°o 1 tham chi·∫øu t·ªõi h√†m create_rnn_model
```

* H√†m tr·∫£ v·ªÅ m·ªôt m√¥ h√¨nh ƒë√£ ƒë∆∞·ª£c compile-phi√™n d·ªãch , s·∫µn s√†ng cho training, v·ªõi h√†m loss l√† ‚Äòmean_squared_error‚Äô v√† tr√¨nh t·ªëi ∆∞u h√≥a ‚Äòadam‚Äô.

### <center> 

```python

def create_lstm_model(input_shape, output_units, units=62):
    model = Sequential() #T·∫°o m√¥ h√¨nh tu·∫ßn t·ª± ƒë·ªÉ c√°c l·ªõp x·∫øp ch·ªìng nhau

    model.add(LSTM(units, input_shape=input_shape, return_sequences=True)) #LSTM layer1 v·ªõi s·ªë ƒë∆°n v·ªã ·∫©n l√† units = 62 ,  tr·∫£ v·ªÅ to√†n b·ªô chu·ªói ƒë·∫ßu ra (t·∫•t c·∫£ c√°c tr·∫°ng th√°i ·∫©n cho m·ªói th·ªùi ƒëi·ªÉm trong chu·ªói ƒë·∫ßu v√†o).
    model.add(Dropout(0.2)) #k·ªπ thu·∫≠t regularization ƒë·ªÉ gi·∫£m overfitting b·∫±ng c√°ch ng·∫´u nhi√™n b·ªè ƒëi m·ªôt t·ª∑ l·ªá ph·∫ßn trƒÉm c·ªßa c√°c ƒë∆°n v·ªã (neuron) trong qu√° tr√¨nh hu·∫•n luy·ªán.
    model.add(LSTM(units, return_sequences=False)) #Th√™m m·ªôt l·ªõp LSTM th·ª© hai  62 ƒë∆°n v·ªã ·∫©n .L·ªõp n√†y ch·ªâ tr·∫£ v·ªÅ tr·∫°ng th√°i ·∫©n cu·ªëi c√πng c·ªßa chu·ªói, thay v√¨ to√†n b·ªô chu·ªói ƒë·∫ßu ra.
    model.add(Dropout(0.2))
    model.add(Dense(output_units)) #L·ªõp Dense (fully connected) v·ªõi  s·ªë ƒë∆°n v·ªã ƒë·∫ßu ra ·ª©ng v·ªõi output_units.L·ªõp n√†y k·∫øt n·ªëi to√†n b·ªô ƒë·∫ßu ra t·ª´ l·ªõp LSTM cu·ªëi c√πng ƒë·∫øn ƒë·∫ßu ra cu·ªëi c√πng c·ªßa m√¥ h√¨nh
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
model_lstm = create_lstm_model
```

## <center> <span style="font-family:Playfair Display"> <span style="color:white;display:fill;border-radius:8px;background-color:#a02933;font-size:100%;font-weight: bold;color: #dbba78;font-style: italic;">„Ö§üé∑ GRU„Ö§</span><a id="1"></a>
* C√≥ hai l·ªõp GRU, m·ªôt l·ªõp cho ph√©p m·∫°ng h·ªçc t·ª´ d·ªØ li·ªáu ·ªü m·ªói b∆∞·ªõc th·ªùi gian, l·ªõp th·ª© hai t√≥m g·ªçn th√¥ng tin cu·ªëi c√πng.
* C√≥ c√°c l·ªõp Dropout gi·ªØa c√°c l·ªõp GRU ƒë·ªÉ ngƒÉn m√¥ h√¨nh h·ªçc thu·ªôc l√≤ng d·ªØ li·ªáu.
* M√¥ h√¨nh d·ª± ƒëo√°n k·∫øt qu·∫£ cu·ªëi c√πng qua m·ªôt l·ªõp Dense v√† ƒë∆∞·ª£c bi√™n d·ªãch s·∫µn s√†ng cho vi·ªác h·ªçc. 


### <center> 

```python
def create_gru_model(input_shape, output_units, units=62):
    model = Sequential()
    model.add(GRU(units, input_shape=input_shape, return_sequences=True))
    model.add(Dropout(0.2))
    model.add(GRU(units, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(output_units))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
model_gru = create_gru_model
```

## <center><span style="font-family:Playfair Display"> <span style="color:white;display:fill;border-radius:8px;background-color:#a02933;font-size:100%;font-weight: bold;color: #dbba78;font-style: italic;">„Ö§üéª Bi_LSTM„Ö§</span><a id="1"></a>  
* S·ª≠ d·ª•ng LSTM hai chi·ªÅu, cho ph√©p m·∫°ng h·ªçc t·ª´ d·ªØ li·ªáu theo c·∫£ hai h∆∞·ªõng, tr∆∞·ªõc sau v√† ng∆∞·ª£c l·∫°i.
* Gi·ªëng nh∆∞ m√¥ h√¨nh GRU, c≈©ng c√≥ hai l·ªõp LSTM v√† c√°c l·ªõp Dropout ƒë·ªÉ tr√°nh h·ªçc thu·ªôc l√≤ng.
* K·∫øt th√∫c b·∫±ng m·ªôt l·ªõp Dense ƒë·ªÉ ƒë∆∞a ra d·ª± ƒëo√°n.
C≈©ng ƒë∆∞·ª£c bi√™n d·ªãch v·ªõi h√†m m·∫•t m√°t v√† optimizer gi·ªëng nh∆∞ GRU.


### <center> 

```python

def create_bi_lstm_model(input_shape, output_units, units=62):
    model = Sequential()
    model.add(Bidirectional(LSTM(units, return_sequences=True), input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(Bidirectional(LSTM(units, return_sequences=False)))
    model.add(Dropout(0.2))
    model.add(Dense(output_units))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
model_bi_lstm = create_bi_lstm_model
```

## <center> <span style="font-family:Playfair Display"> <span style="color:white;display:fill;border-radius:8px;background-color:#a02933;font-size:100%;font-weight: bold;color: #dbba78;font-style: italic;">„Ö§üéπ T·∫°o KerasRegressor„Ö§</span><a id="1"></a>   
L√† 1 tr√¨nh bao b·ªçc c·ªßa th∆∞ vi·ªán scikeras , gi√∫p t√≠ch h·ª£p c√°c m√¥ h√¨nh c·ªßa Keras v·ªõi c√°c c√¥ng c·ª• t·ª´ scikit-learn nh∆∞ c√°c b·ªô t√¨m ki·∫øm GridSearchCV,RandomizeSearchCV, BayesSearchCV gi√∫p t√¨m ra c√°c b·ªô tham s·ªë t·ªët nh·∫•t cho m√¥ h√¨nh

# <center> <span style="font-family:Playfair Display"> <span style="color:white;display:fill;border-radius:8px;background-color:#a02933;font-size:100%;font-weight: bold;color: #dbba78;font-style: italic;">   üçªHu·∫•n luy·ªán m√¥ h√¨nh    </span><a id="1"></a> 

```python
param_dist = {
                'units': range(1, 101),
                'epochs': range(1, 101),
                'batch_size': [1, 2, 4, 8, 16]
                }
def train_model(model, trainX, trainY, validation_data, epochs, batch_size):
    model.fit(trainX, trainY, epochs=epochs, batch_size=batch_size, 
              validation_data=validation_data, verbose=2)

def train_and_time(model, trainX, trainY, validation_data, epochs, batch_size):
    start_time = time.time()
    train_model(model, trainX, trainY, validation_data, epochs, batch_size)
    end_time = time.time()
    training_time = end_time - start_time
    return training_time

def optimize_with_randomized_search(model, param_dist, X, y):
    random_search  = RandomizedSearchCV(model, param_distributions=param_dist, cv=3, n_iter=10, n_jobs=-1,
                                        scoring='neg_mean_squared_error')
    random_search.fit(X, y)
    results = random_search.cv_results_
    best_params = random_search.best_params_
    return results, best_params

def optimize_with_bayesian_optimization(model, param_space, X, y):
    bayes_search = BayesSearchCV(model, search_spaces=param_space, n_iter=10, cv=3, n_jobs=-1, scoring='neg_mean_squared_error')
    bayes_search.fit(X, y)
    results = bayes_search.cv_results_
    best_params = bayes_search.best_params_
    return results,best_params
```
