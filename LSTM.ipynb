```python
def preprocess_data(data, scaling_method, min_max_range=None, zero_mean_range=None):
    data = data.astype('float32')
    if scaling_method == "min-max":
        if min_max_range is None:
            scaler = MinMaxScaler(feature_range=(0, 1))
        else:
            scaler = MinMaxScaler(feature_range=min_max_range)
    elif scaling_method == "zero-mean":
        if zero_mean_range is None:
            scaler = StandardScaler()
        else:
            zero_mean_range = bool(zero_mean_range)
            scaler = StandardScaler(with_mean=zero_mean_range)
    elif scaling_method == "chưa chuẩn hóa":
        scaler = None  
    else:
        raise ValueError("Phương pháp chuẩn hóa không hợp lệ. Chọn 'zero-min', 'min-max' hoặc 'chưa chuẩn hóa'.")
    if scaler is not None:
        scaled_data = scaler.fit_transform(data)
    else:
        scaled_data = data
    return data, scaled_data, scaler
```

Vì đây là bài toán time series với dữ liệu được sắp xếp theo thời gian nên ta cần chia các tập dữ liệu train ,validation, test theo thứ tự thời gian với các phần dữ liệu độc lập nhau.

```python
def train_test_split(data, train_size, validation_size, test_size):
    total_size = len(data)
    train_size = int(total_size * train_size)
    validation_size = int(total_size * validation_size)
    test_size = total_size - train_size - validation_size
    train, validation, test = (
        data[:train_size, :],
        data[train_size:train_size + validation_size, :],
        data[train_size + validation_size:, :]
    )
    return train, validation, test,train_size,validation_size
train_size = 0.6
validation_size = 0.2
test_size = 0.2
train, validation, test , train_size, validation_size= train_test_split(data, train_size, validation_size, test_size)
train
print(test.shape)
```

# <center> <span style="font-family:Playfair Display"> <span style="color:white;display:fill;border-radius:8px;background-color:#a02933;font-size:100%;font-weight: bold;color: #dbba78;font-style: italic;">ㅤ🔢 Tạo ma trận dữ liệu đầu vào và đầu ra cho các mô hình time series LSTM gồm  </span><a id="1"></a>  
* x là các giá trị đầu vào tức T-1 , T-2 ,T-lookback
* y là các giá trị đầu ra tức T, T+1 , T+ look_forward </br>
</br>


```python

def create_dataset(dataset, look_back=1, look_forward=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back - look_forward + 1):
        # Trích xuất dãy đầu vào và giá trị mục tiêu
        seq_x = dataset[i:(i + look_back), 0] #T-1,T-2,T-3,..T-lookback
        target_y = dataset[i + look_back:i + look_back + look_forward, 0] #T,T+1,..,T+look_forward
        # Kiểm tra giá trị NaN trong dãy đầu vào và giá trị mục tiêu
        if not np.isnan(seq_x).any() and not np.isnan(target_y).any():
            dataX.append(seq_x)
            dataY.append(target_y)
    return np.array(dataX), np.array(dataY)
```

Các mạng nơ-ron truyền thống, như CNN, không hiệu quả trong việc xử lý dữ liệu có thứ tự thời gian hoặc chuỗi vì không thể duy trì thông tin từ các bước trước

## <center> <span style="font-family:Playfair Display"> <span style="color:white;display:fill;border-radius:8px;background-color:#a02933;font-size:100%;font-weight: bold;color: #dbba78;font-style: italic;">ㅤ🎸 Simple RNNㅤ</span><a id="1"></a>  
* Hai lớp RNN kết hợp với lớp Dropout để kiểm soát mức điều chuẩn hóa ,giảm overfitting và cải thiện khả năng tổng quát bằng cách tắt 1 số nút  . </br>
* Lớp đầu tiên với ‘return_sequences=True’ giúp giữ lại thông tin chuỗi đầu ra, trong khi lớp thứ hai không cần giữ lại chuỗi đầu ra để tập trung vào thông tin quan trọng. </br>
 

### <center>  </center>

```python

def create_rnn_model(input_shape, output_units, units=62):
    model = Sequential()
    model.add(SimpleRNN(units, input_shape=input_shape, return_sequences=True)) 
    model.add(Dropout(0.2))
    model.add(SimpleRNN(units, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(output_units))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
model_rnn = create_rnn_model #Tạo 1 tham chiếu tới hàm create_rnn_model
```

## <center> <span style="font-family:Playfair Display"> <span style="color:white;display:fill;border-radius:8px;background-color:#a02933;font-size:100%;font-weight: bold;color: #dbba78;font-style: italic;">ㅤ🎺 LSTMㅤ</span><a id="1"></a> 
 * Hai lớp LSTM kết hợp với lớp Dropout để kiểm soát mức điều chuẩn hóa,giảm overfitting và cải thiện khả năng tổng quát bằng cách tắt 1 số nút .</br> 
 * Lớp đầu tiên với ‘return_sequences=True’ giúp giữ lại thông tin chuỗi đầu ra, trong khi lớp thứ hai không cần giữ lại chuỗi đầu ra để tập trung vào thông tin quan trọng. </br>
* Hàm trả về một mô hình đã được compile-phiên dịch , sẵn sàng cho training, với hàm loss là ‘mean_squared_error’ và trình tối ưu hóa ‘adam’.

### <center> 

```python

def create_lstm_model(input_shape, output_units, units=62):
    model = Sequential() #Tạo mô hình tuần tự để các lớp xếp chồng nhau

    model.add(LSTM(units, input_shape=input_shape, return_sequences=True)) #LSTM layer1 với số đơn vị ẩn là units = 62 ,  trả về toàn bộ chuỗi đầu ra (tất cả các trạng thái ẩn cho mỗi thời điểm trong chuỗi đầu vào).
    model.add(Dropout(0.2)) #kỹ thuật regularization để giảm overfitting bằng cách ngẫu nhiên bỏ đi một tỷ lệ phần trăm của các đơn vị (neuron) trong quá trình huấn luyện.
    model.add(LSTM(units, return_sequences=False)) #Thêm một lớp LSTM thứ hai  62 đơn vị ẩn .Lớp này chỉ trả về trạng thái ẩn cuối cùng của chuỗi, thay vì toàn bộ chuỗi đầu ra.
    model.add(Dropout(0.2))
    model.add(Dense(output_units)) #Lớp Dense (fully connected) với  số đơn vị đầu ra ứng với output_units.Lớp này kết nối toàn bộ đầu ra từ lớp LSTM cuối cùng đến đầu ra cuối cùng của mô hình
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
model_lstm = create_lstm_model
```

## <center> <span style="font-family:Playfair Display"> <span style="color:white;display:fill;border-radius:8px;background-color:#a02933;font-size:100%;font-weight: bold;color: #dbba78;font-style: italic;">ㅤ🎷 GRUㅤ</span><a id="1"></a>
* Có hai lớp GRU, một lớp cho phép mạng học từ dữ liệu ở mỗi bước thời gian, lớp thứ hai tóm gọn thông tin cuối cùng.
* Có các lớp Dropout giữa các lớp GRU để ngăn mô hình học thuộc lòng dữ liệu.
* Mô hình dự đoán kết quả cuối cùng qua một lớp Dense và được biên dịch sẵn sàng cho việc học. 


### <center> 

```python
def create_gru_model(input_shape, output_units, units=62):
    model = Sequential()
    model.add(GRU(units, input_shape=input_shape, return_sequences=True))
    model.add(Dropout(0.2))
    model.add(GRU(units, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(output_units))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
model_gru = create_gru_model
```

## <center><span style="font-family:Playfair Display"> <span style="color:white;display:fill;border-radius:8px;background-color:#a02933;font-size:100%;font-weight: bold;color: #dbba78;font-style: italic;">ㅤ🎻 Bi_LSTMㅤ</span><a id="1"></a>  
* Sử dụng LSTM hai chiều, cho phép mạng học từ dữ liệu theo cả hai hướng, trước sau và ngược lại.
* Giống như mô hình GRU, cũng có hai lớp LSTM và các lớp Dropout để tránh học thuộc lòng.
* Kết thúc bằng một lớp Dense để đưa ra dự đoán.
Cũng được biên dịch với hàm mất mát và optimizer giống như GRU.


### <center> 

```python

def create_bi_lstm_model(input_shape, output_units, units=62):
    model = Sequential()
    model.add(Bidirectional(LSTM(units, return_sequences=True), input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(Bidirectional(LSTM(units, return_sequences=False)))
    model.add(Dropout(0.2))
    model.add(Dense(output_units))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
model_bi_lstm = create_bi_lstm_model
```

## <center> <span style="font-family:Playfair Display"> <span style="color:white;display:fill;border-radius:8px;background-color:#a02933;font-size:100%;font-weight: bold;color: #dbba78;font-style: italic;">ㅤ🎹 Tạo KerasRegressorㅤ</span><a id="1"></a>   
Là 1 trình bao bọc của thư viện scikeras , giúp tích hợp các mô hình của Keras với các công cụ từ scikit-learn như các bộ tìm kiếm GridSearchCV,RandomizeSearchCV, BayesSearchCV giúp tìm ra các bộ tham số tốt nhất cho mô hình

```python
kr_LSTM = KerasRegressor(model=model_lstm, input_shape=input_shape, output_units=look_forward, units=62)
#Khi sử dụng kr_LSTM trong một quá trình như tìm kiếm tham số hoặc huấn luyện, KerasRegressor sẽ tạo mô hình bằng cách gọi hàm create_lstm_model với các tham số đã được cung cấp.
kr_RNN = KerasRegressor(model=model_rnn, input_shape=input_shape, output_units=look_forward, units=62)
kr_bi_LSTM = KerasRegressor(model=model_bi_lstm, input_shape=input_shape, output_units=look_forward, units=62)
kr_GRU = KerasRegressor(model=model_gru, input_shape=input_shape, output_units=look_forward, units=62)
```

# <center> <span style="font-family:Playfair Display"> <span style="color:white;display:fill;border-radius:8px;background-color:#a02933;font-size:100%;font-weight: bold;color: #dbba78;font-style: italic;">   🍻Huấn luyện mô hình    </span><a id="1"></a> 

```python
param_dist = {
                'units': range(1, 101),
                'epochs': range(1, 101),
                'batch_size': [1, 2, 4, 8, 16]
                }
def train_model(model, trainX, trainY, validation_data, epochs, batch_size):
    model.fit(trainX, trainY, epochs=epochs, batch_size=batch_size, 
              validation_data=validation_data, verbose=2)

def train_and_time(model, trainX, trainY, validation_data, epochs, batch_size):
    start_time = time.time()
    train_model(model, trainX, trainY, validation_data, epochs, batch_size)
    end_time = time.time()
    training_time = end_time - start_time
    return training_time

def optimize_with_randomized_search(model, param_dist, X, y):
    random_search  = RandomizedSearchCV(model, param_distributions=param_dist, cv=3, n_iter=10, n_jobs=-1,
                                        scoring='neg_mean_squared_error')
    random_search.fit(X, y)
    results = random_search.cv_results_
    best_params = random_search.best_params_
    return results, best_params

def optimize_with_bayesian_optimization(model, param_space, X, y):
    bayes_search = BayesSearchCV(model, search_spaces=param_space, n_iter=10, cv=3, n_jobs=-1, scoring='neg_mean_squared_error')
    bayes_search.fit(X, y)
    results = bayes_search.cv_results_
    best_params = bayes_search.best_params_
    return results,best_params
```

```python
#Predict
def predict(model, dataX):
    predictions = model.predict(dataX)
    return predictions
#Rescale
def rescale_data(data, scaler):
    return scaler.inverse_transform(data)
```

```python

def evaluate_model(y_true, y_pred):
   
    r2 = r2_score(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    '''
    print(f"RMSE: {rmse:.6f}")
    print(f"R^2: {r2:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"MSE: {mse:.4f}")
    '''
    return rmse,r2,mae,mse
```
